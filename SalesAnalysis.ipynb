```python
df['order_date'] = pd.to_datetime(
    df['order_date'],
    format='mixed',
    dayfirst=True
)

df['ship_date'] = pd.to_datetime(
    df['ship_date'],
    format='mixed',
    dayfirst=True
)
```

```python
df.info()
```

Feature engineering

```python
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days
df['order_month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()
df['quarter'] = 'Q' + df['order_date'].dt.quarter.astype(str)

df.quarter.head()

```

```python
print(df['sales'].dtype)
df['profit_margin'] = (df['profit'] / df['sales']) * 100

df.profit_margin.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
plt.figure(figsize=(8,5))
sns.histplot(df['sales'], bins=30, kde=True)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['profit'], bins=30, kde=True)

plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='category', data=df)

plt.title("Orders by Category")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='segment', data=df)

plt.title("Orders by Segment")

plt.show()
```

Bivariate Analysis (EDA)

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')

plt.show()
```

```python
month_sales = df.groupby('month_name')['sales'].sum()

month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

month_sales = month_sales.reindex(month_order)

plt.figure(figsize=(12,5))
month_sales.plot(kind='line', marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

Finding Outliers

```python
import matplotlib.pyplot as plt

numerical_columns = ['sales', 'profit', 'quantity', 'shipping_cost', 'profit_margin']

for col in numerical_columns:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col], vert=False)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower Limit:", lower)
print("Upper Limit:", upper)
```

```python
sales_outliers = df[(df['sales'] < lower) | (df['sales'] > upper)]

print("Number of Sales Outliers:", len(sales_outliers))
```

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower) | (df[column] > upper)]

    print(f"{column}")
    print("Outliers:", len(outliers))
    print("-"*30)

detect_outliers(df, 'sales')
detect_outliers(df, 'profit')
detect_outliers(df, 'quantity')
detect_outliers(df, 'shipping_cost')
detect_outliers(df, 'profit_margin')
```

```python
Q1 = df['quantity'].quantile(0.25)
Q3 = df['quantity'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['quantity'] >= lower) & (df['quantity'] <= upper)]

print("Dataset Shape after removing Quantity outliers:", df.shape)
```

```python
Q1 = df['profit_margin'].quantile(0.25)
Q3 = df['profit_margin'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['profit_margin'] >= lower) & (df['profit_margin'] <= upper)]

print("Final Dataset Shape:", df.shape)
```

```python
detect_outliers(df, 'quantity')
detect_outliers(df, 'profit_margin')
```

```python
Q1 = df['profit_margin'].quantile(0.25)
Q3 = df['profit_margin'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['profit_margin'] >= lower) & (df['profit_margin'] <= upper)]

print("Dataset Shape:", df.shape)
detect_outliers(df, 'profit_margin')
```

```python
category_summary = df.groupby('category')[['sales', 'profit']].sum()

category_summary.plot(kind='bar', figsize=(8,5))

plt.title('Sales and Profit by Category')
plt.xlabel('Category')
plt.ylabel('Amount')

plt.show()
```
